```python
legs=int(input("Enter number of legs: "))
heads=int(input("Enter Total Heads: "))
 flag=False
for cows in range(0,heads+1):
  hens=heads-cows
  cal_legs=cows*4,hens*2
  if(cal_legs==legs):
    flag=True
    print("Number of Cows",cows)
    print("Number of Hens",hens)
    break
if flag:
  print("cows",cows)
  print("hens",hens)
else:
  print("No solution")
```

```python
def is_perfect_square(num):
  if num < 0:
    return False
  a = num**0.5
  return a % 1 == 0

for num in range(1, 100):
  if is_perfect_square(num):
    print(f"{num} is a perfect square")
```

```python
a=input("Enter your password: ")
up=0
lw=0
num=0
sp=0
if len(a)>7:
  for i in a:
      if i.isupper():
        up=up+1
      elif i.islower():
        lw=lw+1
      elif i.isnumeric():
        num=num+1
      else:
        sp=sp+1
  if up>0 and lw>0 and num>0 and sp>0:
    print("Strong Password")
  else:
    print("Weak Password")
else:
  print("Weak Password due to less characters")
```

```python
a=input("Enter your palindrome: ")
b=a.lower()
c="".join(i for i in b if i.isalnum())

if c == c[::-1]:
    print(f"'{a}' is a palindrome.")
else:
    print(f"'{a}' is not a palindrome.")
```
